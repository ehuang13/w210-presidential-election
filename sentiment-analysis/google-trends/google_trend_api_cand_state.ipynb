```python
import pandas as pd
from os.path import join, exists
from os import makedirs
from pytrends.request import TrendReq
```

```python
pytrend = TrendReq(hl='en-US', tz=360)
```

```python
def get_and_save_trends(start_dt, end_dt, keyword, swing_states, path):
    
    makedirs(path, exist_ok=True)
    
    for st in swing_states:
        print(st)
        pytrend.build_payload(
             kw_list=keyword,
             cat=0,
             timeframe= start_dt + ' ' + end_dt,
             geo=st,
             gprop='')
        
        data_trend = pytrend.interest_over_time()
    
        data_trend.to_csv(join(path, st + '_trends.csv'), encoding='utf_8_sig')
```

```python
trump_20_search_terms = ['trump immigration', 'trump economy', 'trump environment', 'trump health care']
biden_20_search_terms = ['biden immigration', 'biden economy', 'biden environment', 'biden health care']
```

```python
start_dt = '2020-05-27'
end_dt = '2020-06-30'
```

```python
for idx, tp in enumerate(topics4):
    search_t = trump_20_search_terms[idx]
    print(tp)
    get_and_save_trends(start_dt, end_dt, [search_t], states, join('cand-state', tp, '2020', 'trump'))
```
