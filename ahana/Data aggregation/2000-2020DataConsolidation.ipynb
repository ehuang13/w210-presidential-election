```python
import pandas as pd
import glob
import os
```

```python
path = os.getcwd()
path
```

```python
#Does not include File35
#https://www2.census.gov/programs-surveys/popest/datasets/2000-2010/intercensal/county/
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame.columns
```

```python
frame.describe()
```

```python
replace_values = { 1 : 2000, 2 : 2001, 3 : 2002 , 4 : 2003, 5 : 2004, 6 : 2005, 7 : 2006, 8 : 2007, 9 : 2008, 10 : 2009 , 11 : 2010, 12 : 2011 , 13 : 2012 }     
frame = frame.replace({"YEAR": replace_values})  
```

```python
frame['AGEGRP'] = frame['AGEGRP'].replace([99],"Total")
frame['AGEGRP'] = frame['AGEGRP'].replace([0,1,2,3,4],"Less than 19")
frame['AGEGRP'] = frame['AGEGRP'].replace([5,6,7,8],"20 to 39")
frame['AGEGRP'] = frame['AGEGRP'].replace([9,10,11,12],"40 to 59")
frame['AGEGRP'] = frame['AGEGRP'].replace([13,14,15,16,17,18],"Above 60")
frame.tail()
```

```python
years=[2000,2004,2008]
frame_year = frame[frame.YEAR.isin(years)]
```
