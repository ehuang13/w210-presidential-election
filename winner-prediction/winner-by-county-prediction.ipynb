```python
def adjust_weights_and_get_acc(df2):

    #df2 = df[['STATE_FIPS', 'STATE', 'WINNING_PARTY', 'PRED_WINNING_PARTY', 'PRED_DEM_VOTES', 'PRED_REP_VOTES']]
    weight = 0.05
    
    for idx, row in df2.iterrows():
        
        if str(row['YEAR']) == '2020':
        
            dem_votes = 0
            rep_votes = 0

            dem_votes = row['DEM_VOTES'] + (row['COUNTY_TOTALVOTES'] * weight)
            rep_votes = row['REP_VOTES'] - (row['COUNTY_TOTALVOTES'] * weight)

            df2.loc[idx,'DEM_VOTES'] = dem_votes
            df2.loc[idx,'REP_VOTES'] = rep_votes

            if dem_votes > rep_votes and row['WINNING_PARTY_BINARY'] == 1:
                print('rep->dem', row['STATE_FIPS'])
                df2.loc[idx,'WINNING_PARTY_BINARY'] = 0
                df2.loc[idx,'HOUSE_WINNING_BINARY'] = 0
                df2.loc[idx,'WINNER_SHIFT'] = 'REP->DEM'
                df2.loc[idx,'DEM_CANDIDATE'] = 'Barack Obama'
            elif dem_votes < rep_votes and row['WINNING_PARTY_BINARY'] == 0:
                print('dem->rep', row['STATE_FIPS'])
                df2.loc[idx,'WINNING_PARTY_BINARY'] = 1
                df2.loc[idx,'HOUSE_WINNING_BINARY'] = 1
                df2.loc[idx,'WINNER_SHIFT'] = 'DEM->REP'
                df2.loc[idx,'DEM_CANDIDATE'] = 'Mitt Romney'
 
   
    return df2
```

```python
data = adjust_weights_and_get_acc(df_20)

data.head(50)

data.to_csv('data-prediction-final/predictions_vs_adjusted_county_2020.csv')
```
