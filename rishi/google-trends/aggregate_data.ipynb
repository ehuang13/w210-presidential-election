```python
import json
import os
import glob
import re
import pandas as pd
from os import makedirs
from os.path import join, exists
from pathlib import Path
from datetime import datetime, date, timedelta
from pathlib import Path
```

```python
def drop_df_columns(df):
    cols = df.columns
    for c in cols:
        if 'isPartial' in c:
            df = df.drop([c], axis=1)
            
    return df
```

```python
def get_and_save_trends(start_dt, end_dt, keyword, swing_states, path, target_path, cand):
    
    makedirs(target_path, exist_ok=True)
    result_df = []

    for st in swing_states:
        file_path = join(path, keyword[0] + '_' + st + '_trends.csv')
        df = pd.read_csv(file_path)
        
        df = drop_df_columns(df)
        
        if len(df) == 0:
            continue
        
        if len(result_df) > 0:
            result_df = pd.merge(df, result_df, how='inner', on=['date', 'date'], 
                                 suffixes=('_' + st + '1', '_y' + st + '2'))
        else:
            result_df = df
       
    print(result_df.columns)
    
    col = result_df.loc[: , result_df.columns[1]: result_df.columns[len(result_df.columns) - 1]]
    
    result_df['score'] = col.mean(axis=1)
    
    result_df_min = result_df[['date', 'score']]
    result_df_min.to_csv(join(target_path, cand + '_trends.csv'))
    
```

```python
sws_16 = ['US', 'US-CO', 'US-FL', 'US-IA', 'US-MI', 'US-MS', 'US-NV', 'US-NH', 'US-OH', 'US-NC', 'US-PA', 'US-VA', 'US-WI']

get_and_save_trends('2016-08-07', '2016-11-08', ['vote trump'], 
                    sws_16, 
                    join('data', 'candidate', '2016', 'trump'), join('data2', 'candidate', '2016', 'trump'), 'trump') 
get_and_save_trends('2016-08-07', '2016-11-08', ['vote hillary'], 
                    sws_16,
                    join('data', 'candidate', '2016', 'hillary'), join('data2', 'candidate', '2016', 'hillary'), 'hillary') 
```
