```python
import pandas as pd
from os.path import join, exists
from os import makedirs
from pytrends.request import TrendReq
```

```python
pytrend = TrendReq(hl='en-US', tz=360)
```

```python
def get_and_save_trends(start_dt, end_dt, keyword, geo, path):
    pytrend.build_payload(
         kw_list=keyword,
         cat=0,
         timeframe= start_dt + ' ' + end_dt,
         geo=geo,
         gprop='')
    data = pytrend.interest_over_time()
    data= data.drop(labels=['isPartial'],axis='columns')
    
    makedirs(path, exist_ok=True)
    data.to_csv(join(path, keyword[0] + '_trends.csv'), encoding='utf_8_sig')
```

```python
get_and_save_trends('2016-08-07', '2016-11-08', ['trump'], 'US', join('data', 'candidate', '2016')) 
```

```python
get_and_save_trends('2016-08-07', '2016-11-08', ['hillary'], 'US', join('data', 'candidate', '2016')) 
```
