```python
import json
import os
import glob
import re
import pandas as pd
from os import makedirs
from os.path import join, exists
from datetime import datetime, date, timedelta
from pathlib import Path

from os import makedirs
from os.path import join, exists
```

```python
def get_election_data(start_date, end_date, data_path):

    df_all = []
    dayrange = range((end_date - start_date).days + 1)
    
    for daycount in dayrange:
        dt = start_date + timedelta(days=daycount)
        datestr = dt.strftime('%Y_%m_%d')
        fname = join(data_path, datestr + '.csv')
        print(fname)
        
        df_tweet = pd.read_csv(fname, lineterminator='\n')
        
        df_tweet = df_tweet[['created_at', 'text']]
        
        if len(df_all) == 0:
            df_all = df_tweet
        else:
            df_all = pd.concat([df_all, df_tweet])
            
    return df_all
```

```python
def combine_and_save_tweet_files(src_path, target_path, start_date, end_date, year, cand):
    df_twt = get_election_data(start_date, end_date, src_path)
    
    os.makedirs(target_path, exist_ok=True)
    
    df_twt.to_csv(join(target_path, cand + '.csv'))
                  
```

```python
def get_date_window_and_cand(yr):
    start_dt= datetime.today()
    end_dt = datetime.today()
    cand1 = ''
    cand2 = ''    
        
    if yr == '2012':
        end_dt = date(2012, 11, 4)
        start_dt = end_dt - timedelta(days=17)
        cand1 = 'obama'
        cand2 = 'romney'
    elif yr == '2016':
        end_dt = date(2016, 11, 6)
        start_dt = end_dt - timedelta(days=29)
        cand1 = 'hillary'
        cand2 = 'trump'
    else:
        end_dt = date(2020, 6, 29)
        start_dt = end_dt - timedelta(days=32)
        cand1 = 'biden'
        cand2 = 'trump'
        
    return start_dt, end_dt, cand1, cand2
```

```python
def agg_all():
    
    for yr in ['2012', '2016', '2020']:
        
        start_dt, end_dt, cand1, cand2 = get_date_window_and_cand(yr)
        
        for tp in ['candidate', 'economy', 'health', 'immigration', 'environment']:
            
            combine_and_save_tweet_files(join('data2','raw', tp, yr, cand1), 
                             join('data2','aggregate',tp,yr), 
                             start_dt, end_dt, yr, cand1)
            
            combine_and_save_tweet_files(join('data2', 'raw', tp, yr, cand2), 
                             join('data2', 'aggregate', tp, yr), 
                             start_dt, end_dt, yr, cand2)
    
```

```python
agg_all()
```
