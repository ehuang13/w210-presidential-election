```python
import pandas as pd
```

```python
pd.set_option('max_columns', None)
```

```python
original_predictions_df = pd.read_csv('../../data/F20_step3_output_724.csv')
historical_df = pd.read_csv("../../data/Data-Jul16/combined_jul16.csv", encoding = "ISO-8859-1")
```

```python
original_predictions_df
```

```python
historical_df
```

```python
#append PRED to all prediction columns
original_predictions_df.rename(columns = {'WINNING_PARTY_BINARY':'PRED_WINNING_PARTY_BINARY', 'REP_VOTES%':'PRED_REP_VOTES%', 'DEM_VOTES%':'PRED_DEM_VOTES%', 'COUNTY_TOTALVOTES':'PRED_COUNTY_TOTALVOTES'}, inplace=True )
```

```python
original_predictions_df
```

```python
#remove unnecessary columns 
predictions_df = original_predictions_df[['YEAR_FIPS','STATE_FIPS', 'COUNTY_FIPS', 'PRED_WINNING_PARTY_BINARY', 'PRED_REP_VOTES%', 'PRED_DEM_VOTES%', 'PRED_COUNTY_TOTALVOTES']]
```

```python
predictions_df
```

```python
merge_df = pd.merge(predictions_df, historical_df, on='YEAR_FIPS')
```

```python
merge_df
```

```python
merge_df = merge_df[['YEAR_FIPS','STATE_FIPS_x', 'COUNTY_FIPS_y', 'PRED_WINNING_PARTY_BINARY', 'WINNING_PARTY_BINARY', 'PRED_COUNTY_TOTALVOTES', 'COUNTY_TOTALVOTES']]
```

```python
merge_df
```

```python
count = 0
match = 0

for index, row in merge_df.iterrows():
    count += 1
    if (row['PRED_WINNING_PARTY_BINARY'] == row['WINNING_PARTY_BINARY']):
        match += 1
```

```python
print('accurately predicted {} out of {} counties; accuracy={:0.3f}'.format(match, count, match/count))
```

```python
merge_df.to_csv("../../data/predictions_vs_actuals_2012.csv")
```
