### Linear regression to preidct total votes by county

```python
X = data_historical.drop(['COUNTY_TOTALVOTES'], axis=1)
y = data_historical[['COUNTY_TOTALVOTES']].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
type(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
```

```python
y_pred = linear_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy score={:.2f}".format(linear_model.score(X_test, y_test) * 100))
```

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df.head(25)
```
