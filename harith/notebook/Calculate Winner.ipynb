```python
def sum_of(x):
    d = 0
    r = 0
    
    for c in x:
        if (c == 'D'):
            d += 1
        else:
            r += 1
    return d, r
```

```python
"""extract how counties prediction for every state"""
visited = {}
electoral_votes = {'D':0, 'R':0 }
for idnex, row in data_2020.iterrows():
    state = int(row['STATE_FIPS'])
    #county = row['COUNTY_FIPS']
    key = state # state + '_' + county
    
    winning_party_binary = row['WINNING_PARTY_BINARY']
    winning_party_chr = 'D' if winning_party_binary == 0 else 'R'
    
    if (visited.get(key) == None):
        visited[state] = list(winning_party_chr)
    else:
        #update existing map
        current_list = visited.get(key)
        current_list.append(winning_party_chr)
```

```python
visited
```

```python
"""determine the winner of every state"""
state_predictions = {}

for key, val in visited.items():
    d, r = sum_of(val)
    winner = 'D' if d > r else 'R'
    state_predictions[key] = winner
```

```python
state_predictions
```

```python
party_electoral = {}
for index, row in df.iterrows():
    state = int(row['STATE_FIPS'])
    votes = int(row['COUNT'])
    party = state_predictions.get(state)
    
    if (party_electoral.get(party) == None):
        party_electoral[party] = votes
    else:
        party_electoral[party] = party_electoral.get(party) + votes
    
```
