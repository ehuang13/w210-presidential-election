### Load and merge senate and house election datasets

```python
# load senate and house datasets
senate = "https://raw.githubusercontent.com/ehuang13/w210-presidential-election/master/data/senate.csv"
house = "https://raw.githubusercontent.com/ehuang13/w210-presidential-election/master/data/house.csv"

senate_df = pd.read_csv(senate, encoding = "ISO-8859-1")
house_df = pd.read_csv(house, encoding = "ISO-8859-1")
```

### Clean up senate election data

```python
# check out senate election dataset
print("Senate data dimensions: {}".format(senate_df.shape))
senate_df.sample(5)
```

```python
# keep only election years data for senate and house elections
election_years = [2000, 2004, 2008, 2012, 2016]

senate_df = senate_df[senate_df["year"].isin(election_years)]
print("Senate data dimensions: {}".format(senate_df.shape))
```

```python
senate_df.head(10)
```

```python
# for each senate election in the state, find the winning party

# group each election by year, state, and party
senate_modified = senate_df.groupby(["year", "state", "state_fips", "party"])["party", "candidatevotes"].max().reset_index()

# sort candidate votes by descending to find the candidates with the most votes
senate_modified = senate_modified.sort_values(by=["year", "state", "state_fips", "candidatevotes"], ascending=False)

# get the party with the most votes by election year and state
senate_modified = senate_modified.groupby(["year", "state", "state_fips"]).first().reset_index()
```

```python
# check out modified senate dataset
senate_modified.head(5)
```

```python
# need to recode winning party to republican or democrat
senate_modified["party"].unique()
```

```python
# code other democratic affiliated parties to democrat
senate_modified = senate_modified.replace(["democrat (not identified on ballot)", "democratic-farmer-labor", 
                                            "democratic-nonpartisan league"], "democrat")

# code indepdent affiliated parties to independent
senate_modified = senate_modified.replace("independent for Maine", "independent")

# filter out independent party elections
senate_modified = senate_modified[senate_modified.party != "independent"]
```

```python
# verify only republican and democrat winning elections remain
senate_modified["party"].unique()
```

```python
# create one hot encoded column for winning party
senate_binary = pd.get_dummies(senate_modified["party"], drop_first = True)
senate_modified["senate_winning_party"] = senate_binary
```

```python
print("Modified senate dataset dimensions: {}".format(senate_modified.shape))
senate_modified.head(5)
```

### Clean up house election data

```python
# checkout house election dataset
print("House data dimensions: {}".format(house_df.shape))
house_df.head(10)
```

```python
# filter for only election years data for senate and house elections
house_df = house_df[house_df["year"].isin(election_years)]
print("House data dimensions: {}".format(house_df.shape))
```

```python
# house election dataset seems to be structured similarly to senate election dataset
house_modified = house_df.groupby(["year", "state", "state_fips", "party"])["party", "candidatevotes"].max().reset_index()

# sort candidate votes by descending to find the candidates with the most votes
house_modified = house_modified.sort_values(by=["year", "state", "state_fips", "candidatevotes"], ascending=False)

# get the party with the most votes by election year and state
house_modified = house_modified.groupby(["year", "state", "state_fips"]).first().reset_index()
```

```python
house_modified.head(5)
```

```python
# need to recode winning party to republican or democrat
house_modified["party"].unique()
```

```python
# code other democratic affiliated parties to democrat
house_modified = house_modified.replace("democratic-farmer-labor", "democrat")

# filter out independent party elections
house_modified = house_modified[house_modified.party != "independent"]
```

```python
# verify only republican and democrat winning elections remain
house_modified["party"].unique()
```

```python
# create one hot encoded column for winning party
house_binary = pd.get_dummies(house_modified["party"], drop_first = True)
house_modified["house_winning_party"] = house_binary
```

```python
print("Modified house dataset dimensions: {}".format(house_modified.shape))
house_modified.head(10)
```

### Merge senate and house election dataset 

```python
# sort candidate votes by descending to find the candidates with the most votes by district
house_senate_merged = pd.merge(house_modified[["year", "state", "state_fips", "house_winning_party"]],
                               senate_modified[["year", "state", "state_fips", "senate_winning_party"]],
                               how="left", left_on=["year", "state_fips"], right_on=["year", "state_fips"])
```
