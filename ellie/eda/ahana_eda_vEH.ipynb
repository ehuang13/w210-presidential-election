```python
data = pd.merge(pop_total,election, on='ID', how='left')
data.tail()
```

```python
data.drop(['year','state','state_po','county'],axis=1)
```

```python
# updated pop df to data df
pop_agegroup = data.groupby(['ID', 'AGEGRP','FIPS','Year','STATE','COUNTY','STNAME','CTYNAME']).sum().drop(['SUMLEV'], axis=1)
```
