```python
# imports
import geopandas as gpd
import pandas as pd
import json
from bokeh.io import show
from bokeh.models import (CDSView, ColorBar, ColumnDataSource,
                          CustomJS, CustomJSFilter, 
                          GeoJSONDataSource, HoverTool,
                          LinearColorMapper, Slider)
from bokeh.layouts import column, row, widgetbox
from bokeh.palettes import brewer
from bokeh.plotting import figure
```

# Plotting States With Bokeh

```python
# read in states shapefile from US Census Bureau
states_usa = gpd.read_file("bokeh/cb_2018_us_state_20m.shp")
states_usa.head()
```

```python
states_usa.shape
```

```python
# try to visualizae Maryland
states_usa.iloc[0]["geometry"]
```

```python
# read in combined dataset
combined_df = pd.read_csv("https://raw.githubusercontent.com/ehuang13/w210-presidential-election/master/data/combined_jun13.csv")
```

```python
combined_df.head()
```

```python
# merge combined dataset with state shapefile
merged_df = states_usa.merge(combined_df, left_on="NAME", right_on="STATE")
merged_df.head()
```

```python
# drop Alaska and Hawaii
merged_df = merged_df.loc[~merged_df["NAME"].isin(["Alaska", "Hawaii"])]
```

```python
# input GeoJSON source that contains features for plotting
geosource = GeoJSONDataSource(geojson = merged_df.to_json())
```

```python
# plot map

# create figure object.
plot = figure(title = 'Test Bokeh Map', 
           plot_height = 600 ,
           plot_width = 950, 
           toolbar_location = 'below',
           tools = "pan, wheel_zoom, box_zoom, reset")
plot.xgrid.grid_line_color = None
plot.ygrid.grid_line_color = None

# add patch renderer to figure.
states = plot.patches('xs','ys', source = geosource,
                   fill_color = None,
                   line_color = "gray", 
                   line_width = 0.25, 
                   fill_alpha = 1)
# create hover tool
plot.add_tools(HoverTool(renderers = [states],
                      tooltips = [('State','@NAME'),
                                ('Votes','@COUNTY_TOTALVOTES')]))

# generates a map on separate tab in browser
show(plot)
```
